Як видно присутні missing values, але дублікатів чи інших проблем немає. Наступним кроком перевіремо кореляції

```python
trainCorrelationMatrix = trainData.corr(numeric_only=True)
trainMask = Numpy.array(trainCorrelationMatrix)
trainMask[Numpy.tril_indices_from(trainMask)] = False

figure, axes = Pyplot.subplots()
figure.set_size_inches(5, 5)
Seaborn.heatmap(trainCorrelationMatrix, cmap = "Reds", mask = trainMask, square = True)
```

```python
testCorrelationMatrix = testData.corr(numeric_only=True)
testMask = Numpy.array(testCorrelationMatrix)
testMask[Numpy.tril_indices_from(testMask)] = False

figure, axes = Pyplot.subplots()
figure.set_size_inches(5, 5)
Seaborn.heatmap(testCorrelationMatrix, cmap = "Reds", mask = testMask, square = True)
```

Надто великих кореляцій немає, але є декілька фіч, що потребують обробки. Наприклад, використання імен, які є унікальними для кожного запису, не буде корисно для кластеризації. Також надто багато унікальних значень у колонці Ticket, і при кодуванні даної категоріальної колонки one hot encoding-ом з'явиться забагато додаткових стовпців і модель буде схильна до перенавчання. Схожа проблема і з колонкою Cabin, але в ній ще й критично великий відсоток missing values. Їхнє заповнення, наприклад, медіаною може призвести до того, що модель буде занадто відхилена від реальних значень. Тож пожертвування даними в колонках є найшвидшим і простішим рішенням.

```python
trainData = trainData.drop("Name", axis=1)
trainData = trainData.drop("Cabin", axis=1)
trainData = trainData.drop("Ticket", axis=1)

testData = testData.drop("Name", axis=1)
testData = testData.drop("Cabin", axis=1)
testData = testData.drop("Ticket", axis=1)
```

Тепер заповнимо медіанами missing values у колонках, в яких їхня кількість відносно невелика



```python
trainData["Embarked"].fillna(value="S", limit=2, inplace=True)
trainData["Age"].fillna(trainData["Age"].mean(), inplace=True)

testData["Age"].fillna(testData["Age"].mean(), inplace=True)
testData["Fare"].fillna(testData["Fare"].mean(), inplace=True)
```

Далі закодуємо категоріальні ознаки. Embarked за допомогою one hot encoding та стать за допомогою label encoding-у



```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

embarked = trainData[["Embarked"]]
oneHotEncoder = OneHotEncoder()
labelEncoder = LabelEncoder()

embarkedEncoded = oneHotEncoder.fit_transform(embarked).toarray()
trainData["Sex"] = labelEncoder.fit_transform(trainData["Sex"])

embarkedEncodedDataFrame = Pandas.DataFrame(embarkedEncoded, columns = oneHotEncoder.get_feature_names_out(["Embarked"]))
trainData = trainData.join(embarkedEncodedDataFrame)
trainData.drop(["Embarked"], axis=1, inplace=True)

trainData.sample(5)
```

```python
embarked = testData[["Embarked"]]

testData["Sex"] = labelEncoder.fit_transform(testData["Sex"])
embarkedEncoded = oneHotEncoder.fit_transform(embarked).toarray()

embarkedEncodedDataFrame = Pandas.DataFrame(embarkedEncoded, columns = oneHotEncoder.get_feature_names_out(["Embarked"]))
testData = testData.join(embarkedEncodedDataFrame)
testData.drop(["Embarked"], axis=1, inplace=True)

testData.sample(5)
```

Тепер до кластеризації

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
trainX = trainData.iloc[:, 1:]
testX = testData.iloc[:, 1:]

scaler = StandardScaler()
scaledTrainX = scaler.fit_transform(trainX)
scaledTestX = scaler.fit_transform(testX)
```

Проведемо кластеризацію трьома різними алгоритмами: DBSCAN, Agglomerative та K-Means Clustering. Протестуэмо їх з різними параметри, епсілон для DBSCAN та n_clusters для інших

```python
dbscan = DBSCAN(eps=3)
dbscan.fit(scaledTrainX)
labels = dbscan.labels_
trainData["Cluster_dbscan_3"] = labels

dbscan = DBSCAN(eps=2)
dbscan.fit(scaledTrainX)
labels = dbscan.labels_
trainData["Cluster_dbscan_2"] = labels

agglomerative = AgglomerativeClustering(n_clusters=3)
agglomerative.fit(scaledTrainX)
labels = agglomerative.labels_
trainData["Cluster_agglomerative_3"] = labels

agglomerative = AgglomerativeClustering(n_clusters=5)
agglomerative.fit(scaledTrainX)
labels = agglomerative.labels_
trainData["Cluster_agglomerative_5"] = labels

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaledTrainX)
labels = kmeans.labels_
trainData["Cluster_kmeans_3"] = labels

kmeans = KMeans(n_clusters=16, random_state=42)
kmeans.fit(scaledTrainX)
labels = kmeans.labels_
trainData["Cluster_kmeans_16"] = labels
```

Для сприйняття 16 кластерів це забагато, отже і параметр eps=2 для DBSCAN не найкращий. Виконаємо ті ж самі дії для тестового датасету з більш-менш оптимальними для візуальної оцінки параметрами



```python
dbscan = DBSCAN(eps=3)
dbscan.fit(scaledTestX)
labels = dbscan.labels_
testData["Cluster_dbscan_3"] = labels

agglomerative = AgglomerativeClustering(n_clusters=3)
agglomerative.fit(scaledTestX)
labels = agglomerative.labels_
testData["Cluster_agglomerative_3"] = labels

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaledTestX)
labels = kmeans.labels_
testData["Cluster_kmeans_3"] = labels
```
